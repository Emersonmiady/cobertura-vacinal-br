## Sobre o sarampo

---

### O que é
O sarampo é uma doença infecciosa grave de origem viral (*Measles morbillivirus*) capaz de provocar diversos sintomas, tais como manchas avermelhadas pelo corpo, febre, tosse, faringite, conjuntivite, etc. A taxa de contágio dessa doença é extremamente alta.



### Transmissão
É espalhada pela tosse, espirros, beijos, pelas gotículas que saem quando se fala e qualquer outra forma de contato com fluidos do nariz e boca de uma pessoa infectada, diretamente ou através de objetos (como copos e talheres).

O vírus se instala na mucosa do nariz e dos seios da face para se reproduzir e depois vai para a corrente sanguínea. A transmissão pode ocorrer entre 4 dias antes e 4 dias depois do aparecimento de manchas vermelhas pelo corpo. Despois do contato com alguém doente, a pessoa pode apresentar os sintomas em média após 10 dias, variando de 7 a 18 dias.

### Sintomas e complicações
#### 1. Período Prodrômico: corresponde ao período de tempo entre os primeiros sintomas da doença e o início dos sinais ou sintomas com base no qual o diagnóstico pode ser estabelecido, alguns dos sintomas possíveis sãoː

- Coriza;
- Mal estar geral;
- Febre alta (40ºC);
- Dor de garganta;
- Infecção no nariz;
- Aversão a luz;
- Conjuntivite;
- Tosse com catarro;
- Dificuldade de ingestão;
- Sinal de Koplik (pequenos pontos brancos rodeados de uma zona vermelha, que se agrupam na mucosa interna das bochechas).

#### 2. Período Exantemático: Ocorre piora dos sintomas do período prodrômico, e as complicações podem incluir:

- Erupções cutâneas que aparecem primeiro na cabeça e "descem" com o tempo para os pés, desaparecem em 7 a 10 dias;
- Secreções aumentadas nas vias respiratórias superiores;
- Elevada produção de muco nos pulmões;
- Voz rouca;
- Faringe e boca inflamadas.

#### 3. Período descamativo: nesse período as manchas escurecem e surge a descamação fina, febre e tosse diminuem sensivelmente.
Possíveis complicações:

- Conjuntivite intensa;
- Pneumonia;
- Infecção no ouvido;
- Diarreia;
- Encefalite;
- raramente pode evoluir para a panencefalite esclerosante subaguda.

### Tratamento
Pacientes com sarampo devem descansar, beber bastante água e sucos, ter uma alimentação saudável rica em vitaminas, limpar os olhos com água morna, tomar antitérmicos caso tenham febre alta e evitar coçar as manchas para não deixar feridas e cicatrizes.

O consumo de vitamina A ajuda a proteger crianças com menos de dois anos de complicações nos olhos e diminui a mortalidade. Beber soro fisiológico ajuda a prevenir desidratação causada pela diarreia e vômito

### Prevenção
A prevenção é feita por vacinas. 

### Histórico no Brasil
- Sarampo passa a ser uma doença de notificação compulsória nacional em 1968;
- Programa de vacinação anti-sarampo criada em 1972;
- Vacina contra o sarampo implantada em 1973, com o Plano Nacional de Imunizações (PNI);
- Maior epidemia de sarampo da década de 80, em 1986. A partir daí, as campanhas foram se massficando em alguns Estados;
- Foi somente com o Plano Nacional de Eliminação do Sarampo, em 1992, que o combate foi definido como prioridade de política de saúde;
- Segunda dose da tríplice viral implantada a partir de junho de 2004;
- Vacina tetraviral, combatendo também a varicela, veio em 2013 para o Brasil.

Antes de começar a análise de fato, quero só informar a [definição](http://tabnet.datasus.gov.br/tabdata/LivroIDB/2edrev/f13.pdf) de Cobertura Vacinal. Este conceito é importante entender para analisarmos os dados futuros.

Definição: Cobertura Vacinal é o percentual de crianças vacinadas com vacinas específicas, em determinado espaço geográfico, no ano
considerado. A fórmula segue abaixo:

- C = "Número de crianças com esquema básico completo na idade alvo para determinado tipo de vacina"

- T = "Número de crianças na idade alvo"

$Cobertura\ Vacinal = \frac{C}{T} \cdot 100$

Em que o número de crianças na idade alvo é calculado com dados do IBGE ou Sinasc.

```
# Retirando a coluna " Total"
br_uf_measles_vc = br_uf_measles_vc.drop(' Total', axis=1)
br_uf_measles_vc.head()
```

A coluna "Ano" é do tipo "object", vamos ver o que está acontecendo.

```
# Vendo os valores únicos do "Ano"
br_uf_measles_vc['Ano'].unique()
```

Vamos retirar as linhas que contém os anos de 1999 (valores estranhos no dataframe), 2020 (dados muito recentes) e do "Total" (não faz sentido para a gente).

```
# Retirando os dados dos anos 1999, 2020 e "Total"
years = ['1999', '2020', 'Total']
br_uf_measles_vc = br_uf_measles_vc.query('Ano not in @years')
br_uf_measles_vc['Ano'] = br_uf_measles_vc['Ano'].astype('int64')
br_uf_measles_vc['Ano'].unique()
```

```
# Verificando dados faltantes
br_uf_measles_vc.isna().sum()
```

Uhuu! Sem nenhum dado faltante!

Antes de iniciar a análise de fato, vou fazer um melt nesse dataframe, criando uma coluna somente com os Estados e outra com seus respectivos valores de cobertura vacinal.

```
# Fazendo um melt no dataframe
br_uf = br_uf_measles_vc.columns[1:]
br_uf_measles_vc = pd.melt(br_uf_measles_vc, id_vars='Ano', 
                           value_vars=br_uf, var_name='uf',
                           value_name='cv').rename({'Ano':'ano'}, axis=1)
br_uf_measles_vc.head()
```

Para facilitar as visualizações, vou separar uma coluna com a respectiva região daquele Estado, e assim ver todos os Estados por região em 5 gráficos.

```
# Separando todas as regiões pela ordem alfabética dos Estados
regions_by_uf = ['Norte', 'Norte', 'Norte', 'Norte', 'Nordeste', 
                 'Nordeste', 'Centro-Oeste', 'Sudeste', 'Centro-Oeste',
                 'Nordeste', 'Sudeste', 'Centro-Oeste', 'Centro-Oeste',
                 'Norte', 'Nordeste', 'Nordeste', 'Nordeste', 'Sul',
                 'Sudeste', 'Nordeste', 'Norte', 'Norte', 'Sul', 'Sul',
                 'Nordeste', 'Sudeste', 'Norte']

# Organizando em ordem alfabética dos Estados
br_uf_measles_vc = br_uf_measles_vc.sort_values(['uf', 'ano'])
br_uf_measles_vc['uf'].unique()
```

```
# Adicionando a nova coluna
br_uf_measles_vc['regiao'] = None

uf = br_uf_measles_vc['uf'].unique()
for index, row in br_uf_measles_vc.iterrows():
  for i in range(len(uf)):
    if row['uf'] == uf[i]:
      br_uf_measles_vc.loc[index, 'regiao'] = regions_by_uf[i]
br_uf_measles_vc.head()
```

```
# Gráficos da cobertura vacinal por Estado, em cada região
regions = br_uf_measles_vc['regiao'].unique()
palletes = ['tab10_r', 'nipy_spectral', 'gist_stern', 'Set1', 'cubehelix']
count = 1

display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 5000})'''))
for region in regions:
  wanted_region = regions[count-1]
  uf_order = br_uf_measles_vc.query('regiao == @wanted_region and ano == "2019"').sort_values('cv', ascending=False)['uf']

  plt.figure(figsize=(15, 9))
  ax = sns.lineplot(x='ano', y='cv', hue='uf', palette=palletes[count-1],
                    data=br_uf_measles_vc.query('regiao == @wanted_region'),
                    hue_order=uf_order)
  plt.xticks(range(2000, 2020, 2), fontsize=12)
  plt.yticks(range(0, 150, 10), fontsize=12)
  plt.xlabel('')
  plt.ylabel('Cobertura vacinal (%)\n', fontsize=14)
  plt.title(f'\nCobertura vacinal da tríplice viral por Unidade Federativa da região {wanted_region}, de 2000 a 2019\n',
            fontsize=16)
  plt.legend(bbox_to_anchor=(1.06, 0.75), loc='upper center', frameon=False,
             fontsize=12)
  plt.axhline(y=95, color='black', linestyle='--')
  ax.text(2019.3, 96.4, '95%', fontsize=12)
  plt.grid(linestyle='--')
  plt.ylim(0, 150)
  sns.despine()
  count += 1
```

Tenho que admitir que alguns gráficos, como o das UF's do Norte e Nordeste, não ficaram tão bons, justamente pela grande quantidade de Estados que existe em cada região.

Mas enfim, quando falamos do surto que houve em SP e RJ no ano de 2019, temos São Paulo com uma cobertura abaixo dos 95%, mas RJ não. Será que existe alguma inconsistência nesses dados?

A resposta é não. De acordo com [esse artigo](https://www.aosfatos.org/noticias/desenhamos-fatos-sobre-o-surto-de-sarampo-no-brasil/) (novamente), a grande maioria das pessoas que pegaram o vírus não estão relacionados com a primeira dose da tríplice viral, e sim com a segunda, lançada em 2004, ou a tetraviral, lançada em 2013. Essas pessoas eram jovens entre 15 a 29 anos, os quais não haviam adquirido estas vacinas ainda... 

Infelizmente as análises da 2ª dose da tríplice viral e a tetraviral ficam fora do escopo deste notebook, não tendo como visualizar esses números em SP e RJ, pois como dito no começo, é um pouco complicado explorarmos tais vacinas.

Bom, voltando a análise das UF's, infelizmente a grande maioria delas apresentaram cobertura vacinal da tríplice viral (1ª dose) abaixo da meta prevista nos últimos anos, o que poderia desencadear alguns surtos de sarampo **sim**.

Abaixo vão todas as UF's que apresentaram uma cobertura vacinal abaixo da meta, em 2019.


```
# Todas as UF's que ficaram com cv menor que a meta de 95%
below_goal = br_uf_measles_vc.query('ano == "2019" and cv < 95').drop('ano', axis=1).sort_values('cv', ascending=False)
display(below_goal)
```

Dessas UF's, temos a seguinte quantidade, por região:

```
# Gráfico dos números de ufs por regiao, com cv abaixo de 95% 
plt.figure(figsize=(15, 9))
sns.countplot(x='regiao', data=below_goal, color='steelblue')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('\nRegiões brasileiras', fontsize=14)
plt.ylabel("Número de UF's\n", fontsize=14)
plt.title("Número de UF's por região brasileira que apresentam cobertura vacinal abaixo de 95%, em 2019\n",
          fontsize=16)
sns.despine()
```

Em porcentagens, dentro da própria região, temos:

```
# Porcentagem das UF's dentro das regiões, com cv abaixo dos 95%
pct_in_region = below_goal['regiao'].value_counts() / br_uf_measles_vc.query('ano == "2019"')['regiao'].value_counts()
pct_in_region = pct_in_region.reset_index().rename({'index':'regiao',
                                                    'regiao':'porcentagem_das_ufs'}, axis=1)
pct_in_region.sort_values('porcentagem_das_ufs', ascending=False).head()
```

### Mas por quê?

#### Falta de equipamentos?

A primeira hipótese que eu tive, para a não vacinação, seria a falta de agulhas e frascos por exemplo. Se o Brasil não está tendo nem o orçamento para gastar com esses equipamentos, então realmente está triste a situação.

Entretanto, foi relatado [nesse artigo](http://cebes.org.br/site/wp-content/uploads/2020/04/200408_nota_tecnica_disoc_61.pdf) que, o gasto, em milhões de reais (R$), com esse tipo de coisa está **aumentando**. Veja o gráfico retirado desta fonte:

imunobiologicos_e_insumos.png

Bom, acho que não é a falta de vacinas que está fazendo as pessoas não se vacinarem.

**Observação:** não encontrei datasets para fazer a análise dos gastos.

#### Horários, redução de clínicas e falta de confiança?

De acordo com [essa notícia](https://noticias.uol.com.br/saude/ultimas-noticias/redacao/2020/09/21/vacinacao-criancas-covid-19-coronavirus-cobertura-vacinal.htm), os fatores podem ser explicados, em parte, pela disponibilidade de horários dos postos de saúde, redução das equipes de vacinação e fechamento de várias clínicas durante o passar dos anos.

> "Os postos atendem em horário comercial, alguns fecham para almoço, não funcionam aos finais de semana e feriados..."

> "E muitos estados reduziram suas equipes de agentes de saúde e fecharam Clínicas da Família, restringindo o acesso à vacina."

Além de tudo isso, a médica da notícia diz que, a "nova geração não conhece o risco das doenças porque o nosso programa as controlou".

> "Eles se perguntam 'por que eu vou tomar uma injeção se estou saudável e não conheço essa doença?'."

Sim, é possível que essas afirmações sejam verdadeiras, mas para que isso seja 100% verdade, precisaria de muitos dados que nem sei como ter acesso... infelizmente fica difícil confirmar alguma coisa.

#### Hesitação vacinal?

A hesitação vacinal, segundo a presidente da sociedade, [nessa notícia](https://www.uol.com.br/vivabem/noticias/redacao/2020/09/02/vacinacao-nao-e-obrigatoria-entenda-por-que-fala-do-presidente-e-perigosa.htm), é caracterizado pelo "atraso em aceitar ou a recusa das vacinas recomendadas quando elas estão disponíveis nos serviços de saúde".

> "Muitas vezes relacionamos só à falta de confiança, mas não é verdade. A baixa adesão é multifatorial. Em muitos países da América Latina, são fatores estruturais, de acesso a informação, não exatamente de medo. Só confiar em vacina não faz você ser vacinado. Tem gente que acaba deixando para outra hora, ou chega no posto e não tem vacina, tem uma rotina cheia de compromissos. Isso pesa muito mais."

Sendo assim, isso se chama, basicamente, a falta de uma "cultura de vacinação", ou preguiça mesmo. 

Realmente, podem ser fatores que levam a não vacinação, mas temos que lembrar que, de acordo com as nossas análises desse notebook, as pessoas já tiveram uma cultura de vacinação no passado, pelo menos no caso do sarampo (2003 e 2014 foram os auges da vacinação da 1ª dose da tríplice viral)... Se for esse o caso, devemos fortificar nossas campanhas de vacinação, para que ampliemos a mensagem de que as **vacinas são extremamente importantes**!

#### Movimentos Anti-vacina?

Por mais que já foi comprovado pela Organização Mundial da Saúde que a tríplice viral não causa autismo, ainda é algo em que muitas pessoas acreditam, e citam em artigos acadêmicos por sinal.

Sobre a história, vou deixar [esse link](https://guiadoestudante.abril.com.br/atualidades/entenda-o-que-e-o-movimento-antivacina/) para se divertirem com o maravilhoso mundo das fake news sobre as vacinas. :)

Infelizmente, não consegui achar dados sobre os movimentos anti-vacina. Gostaria muito de verificar se houve alguma relação entre as postagens falsas e a queda da cobertura vacinal... Porém, achei um [artigo](https://sbim.org.br/noticias/1139-sete-a-cada-dez-brasileiros-acreditam-em-informacoes-falsas-sobre-vacinacao) bem interessante, que fala um pouco sobre como as mídias sociais poderiam interferir na vacinação.

Enfim, as fake news são extremamente perigosas, ainda mais quando estamos falando de prevenção contra certas doenças, e sim, pode ser considerado um fator que abaixe a cobertura vacinal de qualquer doença, em vários lugares no mundo.
